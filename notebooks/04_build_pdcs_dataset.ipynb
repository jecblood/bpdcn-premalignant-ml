```python
from pathlib import Path
import pandas as pd

pdcs = []

for file in Path("data/raw").glob("*metadata*.gz"):

    df = pd.read_csv(file, sep="\t")

    pdc = df[df["CellType"] == "pDC"].copy()

    pdc["sample"] = file.name

    pdcs.append(pdc)

pdcs = pd.concat(pdcs)

print(pdcs.shape)
```

```python
pdcs["is_malignant"].value_counts()
```

```python
pdcs["State"] = pdcs["is_malignant"]
```

```python
pdcs.to_csv(
    "data/processed/pdc_metadata.csv",
    index=False
)
```
