```python
meta_cells = set(pt10_meta["cell"])

count_cells = set(pt10_counts.columns)

print("metadata cells:", len(meta_cells))
print("count cells:", len(count_cells))
print("matching:", len(meta_cells & count_cells))
```

```python
def validate_sample(meta, counts):

    meta_cells = set(meta["cell"])

    count_cells = set(counts.columns)

    print("metadata:", len(meta_cells))
    print("counts:", len(count_cells))
    print("intersection:", len(meta_cells & count_cells))

    return len(meta_cells & count_cells)
```

```python
validate_sample(
    pt10_meta,
    pt10_counts
)
```
