```python
from pathlib import Path
import pandas as pd

summary = []

for file in Path("data/raw").glob("*metadata*.gz"):

    df = pd.read_csv(file, sep="\t")

    counts = df["is_malignant"].value_counts()

    summary.append({
        "sample": file.name,
        "healthy": counts.get("Healthy", 0),
        "premalignant": counts.get("Premalignant", 0),
        "malignant": counts.get("Malignant", 0),
        "other": counts.get("Other", 0),
        "total_cells": len(df)
    })

summary_df = pd.DataFrame(summary)

summary_df.sort_values(
    "malignant",
    ascending=False
)
```

```python
summary_df.to_csv(
    "results/metadata_census.csv",
    index=False
)
```
